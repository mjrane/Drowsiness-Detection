# #Face Detection

```python
import cv2

'''
Haar-like features compare the brightness of different rectangular regions within an 
image to quickly extract basic visual patterns for object detection.
'''
#CascadeClassifier method in cv2 module supports the loading of haar-cascade XML files
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml') #specify file paths 
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')

cap = cv2.VideoCapture('video 2.mp4')

while True:
    ret, img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #RGB to Grayscale
    
    '''cv2.COLOR_BGR2GRAY throws errors: 
    error: OpenCV(4.8.0) /io/opencv/modules/imgproc/src/color.cpp:182: error: (-215:Assertion failed) !_src.empty() in function 'cvtColor'
    This is cause the file location is not mentioned correctly
    '''
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    '''
    The detectMultiScale method detects faces by sliding windows at 
    different scales and grouping similar detections, returning bounding boxes with parameters for customization.
    
    It has two input values.
    Scale Factor: Controls how much the image is resized at each step, allowing detection of faces at different sizes.
    Min Neighbors: Determines how many neighboring detections are required to consider a region a face, reducing false positives.'''

    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),1)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        
        eyes = eye_cascade.detectMultiScale(roi_gray)
        print("Eyes", eyes)
        if eyes is ():
            print("eyes closed")
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color,(ex,ey),
                         (ex+ew,ey+eh),(0,255,0),2)
    cv2.imshow('img',img)
    k = cv2.waitKey(1) & 0xff
    if k == 27:
        break

cap.release()
cv2.destroyAllWindows()

```

# Drowsiness Detection using pygame

```python
import cv2
from functools import wraps
from pygame import mixer
import time

lastsave = 0


def counter(func):
    @wraps(func)
    def tmp(*args, **kwargs):
        tmp.count += 1
        global lastsave
        if time.time() - lastsave > 3:
            # this is in seconds, so 5 minutes = 300 seconds
            lastsave = time.time()
            tmp.count = 0
        return func(*args, **kwargs)
    tmp.count = 0
    return tmp


face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')  #loads XML Files

eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')

cap = cv2.VideoCapture(0) #instead of 0 we can also add files(mp4/jpg)
                          # 0 reads camera input


@counter
def closed():
  print ("Eye Closed")


def openeye():
  print ("Eye is Open")



def sound():
    mixer.init()
    mixer.music.load('alarm.mp3')
    mixer.music.play()

while 1:
    ret, img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
        roi_gray = gray[y:y + h, x:x + w]
        roi_color = img[y:y + h, x:x + w]

        eyes = eye_cascade.detectMultiScale(roi_gray)

        if eyes is not ():
            for (ex, ey, ew, eh) in eyes:
                cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)
                openeye()
        else:
           closed()
           if closed.count == 3:
               print ("driver is sleeping")
               sound()

    cv2.imshow('img', img)
    k = cv2.waitKey(1) & 0xff #wait key takes two values 0&1 for delay specification.#instead of 0xff we can also write ord(q)
    if k==27:
        break


cap.release()
cv2.destroyAllWindows()
```
